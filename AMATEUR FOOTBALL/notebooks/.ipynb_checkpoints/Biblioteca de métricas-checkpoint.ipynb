```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Simulamos 20 partidos y cmo evoluciona el ELO de un jugador
partidos = np.arange(1, 21)
elo_inicial = 1500
elo = [elo_inicial]

# Simulamos subidas y bajadas de ELO
np.random.seed(42)
cambios = np.random.randint(-20, 30, size=20)

for cambio in cambios:
    nuevo_elo = elo[-1] + cambio
    elo.append(nuevo_elo)

elo = elo[1:]  # quitamos el ELO inicial

# Crear grfico
plt.figure(figsize=(10, 5))
plt.plot(partidos, elo, marker='o', linestyle='-', color='blue', label='ELO jugador')
plt.axhline(y=elo_inicial, color='gray', linestyle='--', label='ELO inicial (1500)')
plt.title('Evolucin del ELO de un jugador en 20 partidos')
plt.xlabel('Partido')
plt.ylabel('ELO')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Crear el campo estilo StatsBomb
pitch = Pitch(pitch_type='statsbomb', pitch_color='white', line_color='black')
fig, ax = pitch.draw(figsize=(16, 10))

# Definir grid del campo: 6 columnas x 4 filas
n_cols, n_rows = 6, 4
x_bins = np.linspace(0, 120, n_cols + 1)
y_bins = np.linspace(0, 80, n_rows + 1)

# Simular datos de duelos ganados
np.random.seed(42)
duelos_totales = np.random.randint(5, 20, size=(n_rows, n_cols))
duelos_ganados = np.random.randint(0, duelos_totales + 1)
porcentaje_ganados = np.round(100 * duelos_ganados / duelos_totales, 1)

# Dibujar zonas con color segn % ganados
for i in range(n_rows):
    for j in range(n_cols):
        x0, x1 = x_bins[j], x_bins[j + 1]
        y0, y1 = y_bins[i], y_bins[i + 1]
        value = porcentaje_ganados[i, j]
        color = plt.cm.RdYlGn(value / 100)
        ax.add_patch(plt.Rectangle((x0, y0), x1 - x0, y1 - y0, color=color, alpha=0.6))
        ax.text((x0 + x1) / 2, (y0 + y1) / 2, f"{value}%", ha='center', va='center', fontsize=10)

#  Flecha indicando direccin de ataque (de izquierda a derecha)
arrow_start = (10, 75)
arrow_end = (40, 75)
ax.annotate('', xy=arrow_end, xytext=arrow_start,
            arrowprops=dict(facecolor='black', shrink=0.05, width=4, headwidth=10))

ax.text(25, 78, 'Direccin de ataque', ha='center', fontsize=12, color='black')

# Ttulo
plt.title("Porcentaje de Duelos Ganados por Zona\n(Direccin de ataque )", fontsize=16)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch

# --------- 1. Simular datos aleatorios de un jugador ------------
np.random.seed(42)
n_shots = 30
df = pd.DataFrame({
    'x': np.random.uniform(85, 105, n_shots),  # distancia desde portera (ms cerca del 105)
    'y': np.random.uniform(15, 85, n_shots),   # ancho del campo
    'xG': np.random.uniform(0.01, 0.5, n_shots),
    'isGoal': np.random.choice([0, 1], n_shots, p=[0.7, 0.3]),
    'body_part': np.random.choice(['left foot', 'right foot', 'head'], n_shots, p=[0.2, 0.6, 0.2])
})

# --------- 2. Crear campo de ftbol vertical ------------
pitch = VerticalPitch(pitch_type='opta', pitch_color='white', line_color='black')
fig, ax = pitch.draw(figsize=(10, 7))

# --------- 3. Dibujar tiros como burbujas dentro del campo ------------
for _, row in df.iterrows():
    color = 'red' if row['isGoal'] == 1 else 'gray'
    pitch.scatter(row['x'], row['y'], ax=ax,
                  s=row['xG'] * 3000,  # Escala de burbuja
                  color=color, alpha=0.7, edgecolors='black', zorder=3)

# --------- 4. Leyenda explicativa sobre tamao de burbuja (xG) ------------
for xg_val, ypos in zip([0.05, 0.2, 0.4], [10, 20, 30]):
    pitch.scatter(110, ypos, ax=ax, s=xg_val * 3000, color='gray', alpha=0.6, edgecolors='black')
    ax.text(112, ypos, f'xG: {xg_val}', va='center', fontsize=9)

ax.text(110, 5, 'Tamao  prob. de gol', fontsize=10, weight='bold')

# --------- 5. Ttulo ------------
ax.set_title("Tiros del jugador - Distribucin aleatoria\nTamao de burbuja segn xG", fontsize=14, weight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Generar datos aleatorios
# -------------------------------
np.random.seed(42)

num_jugadores = 20
jugadores = [f'Jugador {i+1}' for i in range(num_jugadores)]

# Simulacin: ms acciones suelen llevar ms prdidas
acciones_tercio = np.random.normal(loc=25, scale=6, size=num_jugadores).clip(min=5)
perdidas_tercio = (acciones_tercio * np.random.normal(loc=0.3, scale=0.1, size=num_jugadores)).clip(min=0)

# Crear DataFrame
df = pd.DataFrame({
    'Jugador': jugadores,
    'Acciones ofensivas en el ltimo tercio': acciones_tercio,
    'Prdidas en el ltimo tercio': perdidas_tercio
})

# -------------------------------
# Grfico de dispersin
# -------------------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Prdidas en el ltimo tercio',
    y='Acciones ofensivas en el ltimo tercio',
    s=100,
    color='darkorange'
)

# Aadir etiquetas con nombres de jugadores
for i in range(df.shape[0]):
    plt.text(
        df['Prdidas en el ltimo tercio'][i] + 0.2,
        df['Acciones ofensivas en el ltimo tercio'][i],
        df['Jugador'][i],
        fontsize=8
    )

plt.title('Riesgo vs Volumen ofensivo en el ltimo tercio')
plt.xlabel('Prdidas en el ltimo tercio (por 90 min)')
plt.ylabel('Acciones ofensivas en el ltimo tercio (por 90 min)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# -------------------------------
# Simulacin de datos
# -------------------------------
np.random.seed(42)

# Simular 200 pases progresivos con coordenadas X, Y (campo de 105x68 m)
# Zonas tpicas para estos pases: medio campo y parte ofensiva
x = np.random.normal(loc=60, scale=15, size=200).clip(0, 105)  # Longitud del campo
y = np.random.normal(loc=34, scale=15, size=200).clip(0, 68)   # Anchura del campo

# -------------------------------
# Dibujar campo y mapa de calor
# -------------------------------
fig, ax = plt.subplots(figsize=(12, 7))

# Dibujar el campo (estilo horizontal)
plt.plot([0, 0], [0, 68], color="black")
plt.plot([0, 105], [68, 68], color="black")
plt.plot([105, 105], [68, 0], color="black")
plt.plot([105, 0], [0, 0], color="black")
plt.plot([52.5, 52.5], [0, 68], color="black", linestyle='--')  # Lnea de medio campo

# Mapa de calor
sns.kdeplot(
    x=x,
    y=y,
    shade=True,
    cmap='Reds',
    bw_adjust=1,
    alpha=0.7,
    ax=ax,
    thresh=0.05,
    clip=((0, 105), (0, 68))
)

# Detalles visuales
plt.title("Mapa de calor: pases progresivos por zona del campo", fontsize=14)
plt.xlim(0, 105)
plt.ylim(0, 68)
plt.xlabel('Longitud del campo (m)')
plt.ylabel('Anchura del campo (m)')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Parmetros de campo y simulacin
# -------------------------------
np.random.seed(42)
n_acciones = 300

# Campo de 105x68
x = np.random.normal(loc=70, scale=15, size=n_acciones).clip(0, 105)
y = np.random.normal(loc=34, scale=20, size=n_acciones).clip(0, 68)

# xT simulado: ms alto cuanto ms cerca del rea
xT = np.clip(np.random.normal(loc=0.05 + (x / 105) * 0.1, scale=0.02), 0, 0.3)

df = pd.DataFrame({'X': x, 'Y': y, 'xT': xT})

# -------------------------------
# Crear cuadrcula de 50 zonas
# -------------------------------
n_x_bins = 10  # dividir largo del campo en 10
n_y_bins = 5   # dividir ancho del campo en 5

df['X_bin'] = pd.cut(df['X'], bins=n_x_bins, labels=False)
df['Y_bin'] = pd.cut(df['Y'], bins=n_y_bins, labels=False)

# Calcular xT promedio o acumulado por zona
heatmap_data = df.groupby(['X_bin', 'Y_bin'])['xT'].mean().unstack().fillna(0)

# -------------------------------
# Dibujar campograma binned
# -------------------------------
fig, ax = plt.subplots(figsize=(12, 7))

# Mostrar los valores en formato rotado
for x_bin in range(n_x_bins):
    for y_bin in range(n_y_bins):
        val = heatmap_data.iloc[x_bin, y_bin]
        plt.text(
            (x_bin + 0.5) * (105 / n_x_bins),
            (y_bin + 0.5) * (68 / n_y_bins),
            f"{val:.2f}",
            ha='center',
            va='center',
            fontsize=8,
            color='black'
        )

# Mostrar heatmap como fondo de color
plt.imshow(
    heatmap_data.T,
    origin='lower',
    extent=[0, 105, 0, 68],
    cmap='YlOrRd',
    alpha=0.8,
    aspect='auto'
)

# Dibujar lneas del campo
for x in np.linspace(0, 105, n_x_bins + 1):
    plt.plot([x, x], [0, 68], color='black', linewidth=0.5)
for y in np.linspace(0, 68, n_y_bins + 1):
    plt.plot([0, 105], [y, y], color='black', linewidth=0.5)

# Etiquetas y formato
plt.title("Mapa zonal de xT (Expected Threat)", fontsize=14)
plt.xlabel("Longitud del campo (m)")
plt.ylabel("Anchura del campo (m)")
plt.xlim(0, 105)
plt.ylim(0, 68)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Crear datos simulados
# -------------------------------
np.random.seed(42)
n_jugadores = 50

# Datos de jugadores aleatorios
jugadores = [f"Jugador_{i+1}" for i in range(n_jugadores)]
duelos_def_90 = np.random.normal(8, 2, n_jugadores).clip(3, 14)
intercepciones_entradas_90 = np.random.normal(6, 1.5, n_jugadores).clip(2, 10)
exito_duelos = np.random.normal(0.6, 0.1, n_jugadores).clip(0.3, 0.9)

df = pd.DataFrame({
    'Jugador': jugadores,
    'Duelos_90': duelos_def_90,
    'Intercepciones_Entradas_90': intercepciones_entradas_90,
    'Exito_Duelos': exito_duelos
})

# Aadir tu jugador manualmente
df.loc[len(df.index)] = ['Mi_Jugador', 10.5, 8.2, 0.78]

# -------------------------------
# Graficar dispersin
# -------------------------------
plt.figure(figsize=(10, 6))

# Otros jugadores
otros = df[df['Jugador'] != 'Mi_Jugador']
plt.scatter(
    otros['Duelos_90'],
    otros['Intercepciones_Entradas_90'],
    c=otros['Exito_Duelos'],
    cmap='coolwarm',
    s=70,
    alpha=0.7,
    label='Resto de jugadores'
)

# Tu jugador destacado
mi_jugador = df[df['Jugador'] == 'Mi_Jugador']
plt.scatter(
    mi_jugador['Duelos_90'],
    mi_jugador['Intercepciones_Entradas_90'],
    color='gold',
    edgecolors='black',
    s=200,
    label='Mi Jugador'
)

# Etiquetas
plt.xlabel('Duelos defensivos por 90 min')
plt.ylabel('Intercepciones + entradas por 90 min')
plt.title('Acciones Defensivas por 90 min  Comparativa')
plt.colorbar(label='% xito en duelos')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Crear datos aleatorios
# -----------------------------
np.random.seed(42)
n_jugadores = 30
jugadores = [f"Jugador_{i+1}" for i in range(n_jugadores)]

# Intercepciones totales
intercepciones = np.random.randint(10, 50, n_jugadores)

# Contraataques generados a partir de intercepciones
contraataques = (intercepciones * np.random.uniform(0.1, 0.7, n_jugadores)).astype(int)

# Calcular ratio
ratio = contraataques / intercepciones

df = pd.DataFrame({
    'Jugador': jugadores,
    'Intercepciones': intercepciones,
    'Contraataques': contraataques,
    'Ratio_Interception_to_Counter': ratio
})

# Aadir tu jugador destacado
df.loc[len(df)] = ['Mi_Jugador', 38, 20, 20 / 38]

# Ordenar para grfico
df_sorted = df.sort_values('Ratio_Interception_to_Counter', ascending=False)

# -----------------------------
# Grfico de barras
# -----------------------------
plt.figure(figsize=(12, 6))
bars = plt.bar(
    df_sorted['Jugador'],
    df_sorted['Ratio_Interception_to_Counter'],
    color=['gold' if name == 'Mi_Jugador' else '#1f77b4' for name in df_sorted['Jugador']]
)

plt.xticks(rotation=75)
plt.ylabel('Interception-to-Counter Ratio')
plt.ylim(0, 1)
plt.title('Transformacin defensiva en ataque: Ratio Intercepcin  Contraataque')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Datos aleatorios
# ---------------------------
np.random.seed(42)
n_jugadores = 20
jugadores = [f"Jugador_{i+1}" for i in range(n_jugadores)]

# Generar turnovers y tiros aleatorios (siempre menos o igual)
high_turnovers = np.random.randint(5, 30, size=n_jugadores)
shots_after_turnover = [np.random.randint(0, ht + 1) for ht in high_turnovers]

df = pd.DataFrame({
    'Jugador': jugadores,
    'High_Turnovers': high_turnovers,
    'Shots_After': shots_after_turnover
})
df['HT_Shot_Ratio_%'] = (df['Shots_After'] / df['High_Turnovers']) * 100

# ---------------------------
# Aadir tu jugador personalizado
# ---------------------------
nuevo_jugador = {
    'Jugador': 'Mi_Jugador',
    'High_Turnovers': 21,
    'Shots_After': 11,
    'HT_Shot_Ratio_%': (11 / 21) * 100
}
df = pd.concat([df, pd.DataFrame([nuevo_jugador])], ignore_index=True)

# ---------------------------
# Grfico de dispersin
# ---------------------------
plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    color = 'gold' if row['Jugador'] == 'Mi_Jugador' else '#1f77b4'
    plt.scatter(row['High_Turnovers'], row['HT_Shot_Ratio_%'], color=color, s=80, edgecolors='black')
    if row['Jugador'] == 'Mi_Jugador':
        plt.text(row['High_Turnovers'] + 0.5, row['HT_Shot_Ratio_%'], row['Jugador'], fontsize=9, fontweight='bold')

plt.xlabel('Recuperaciones Altas (High Turnovers)')
plt.ylabel('% que terminan en Tiro')
plt.title('Eficiencia Ofensiva Tras Presin Alta  High Turnover to Shot Ratio por Jugador')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
