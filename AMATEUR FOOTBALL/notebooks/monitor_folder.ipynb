# SCRIPT MONITOREO DE CARPETA DATA, LANZAMIENTOS DE SCRIPTS

```python
import os
import time
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
import subprocess

DATA_FOLDER = '../data/'

class NewFileHandler(FileSystemEventHandler):
    def on_created(self, event):
        if event.is_directory:
            return
        file_path = event.src_path
        if file_path.endswith('.xlsx') and not os.path.basename(file_path).startswith('~$'):
            print(f"Nuevo archivo detectado: {file_path}")
            
            # Ejecutar script de limpieza y capturar su salida
            print("Ejecutando script de limpieza...")
            result = subprocess.run(['python', 'clean_data.py', file_path], capture_output=True, text=True)

            # Buscar la ruta del archivo limpio en la salida
            output_path = None
            for line in result.stdout.splitlines():
                if "Ruta completa del archivo limpio:" in line:
                    output_path = line.split(":", 1)[1].strip()
                    break

            if output_path:
                print("Subiendo datos a Azure...")
                subprocess.run(['python', 'upload_to_azure.py', output_path])
            else:
                print("No se encontr√≥ la ruta del archivo limpio. No se subieron datos.")

if __name__ == "__main__":
    event_handler = NewFileHandler()
    observer = Observer()
    observer.schedule(event_handler, DATA_FOLDER, recursive=False)
    observer.start()
    print(f"Monitoreando la carpeta {DATA_FOLDER} para nuevos archivos Excel...")
    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()
```
