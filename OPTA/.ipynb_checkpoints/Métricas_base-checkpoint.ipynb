## Expected Threat

```python
#Nos traemos el modelo que vamos a usar ya preentrenado con los partidos de la liga 2017-2018 de la libreria propia
url_grid = "https://karun.in/blog/data/open_xt_12x8_v1.json"
xT_model = xthreat.load_model(url_grid)
```

```python
#Nuevo dataframe con todas las acciones que dan continuidad al juego (pases y tiros en este primer modelo)
df_xT = df[(df['isShot'] == True) | (df['type_displayName'] == 'Pass')]
df_xT
```

```python
#funciones para pasar del formato del partido al formato spadl.
def parseo_spadl(partido):
    partido.rename(columns={
        "id":"game_id",
        "eventId":"event_id",
        "teamId":"team_id",
        "x": "start_x",
        "y": "start_y",
        "endX":"end_x",
        "endY":"end_y",
        "period_value":"value",
        "period_displayName":"displayName",
        "type_value": "type_id",
        "type_displayName": "type_name",
        "outcomeType_value": "result_id",
        "outcomeType_displayName": "result_name",
        "playerId": "player_id"
    }, inplace=True)
    partido['type_name'] = np.where(partido['isShot'] == 1, 'shot', 'pass')
    partido['type_id'] = np.where(partido['isShot'] == 1, 11, 0)
    partido['isGoal'] = partido.isGoal.fillna(0)
    partido['result_name'] = np.where((partido['type_name'] == 'shot') & (partido['isGoal'] != 1), 'fail', np.where((partido['result_id'] != 1) & (partido['type_name'] == 'pass'), 'Unsuccessful', 'Successful'))
    partido[['end_x','end_y']] = partido[['end_x', 'end_y']].fillna(0)
    partido['start_x'] = partido['start_x'] * 1.05
    partido['start_y'] = partido['start_y'] * 0.68
    partido['end_x'] = partido['end_x'] * 1.05
    partido['end_y'] = partido['end_y'] * 0.68
```

```python
print(df[["minute", "second", "time_seconds"]].head(50))
```

```python
#Se aplica la funcion anteriormente definida y se aplica para conseguir el dataframe en formatos SPADL
parseo_spadl(df_xT)
columnas_deseadas = [
    'game_id', 'event_id', 'minute', 'second', 'team_id',
    'start_x', 'start_y', 'player_id', 'end_x', 'end_y',
    'isShot', 'isGoal', 'value', 'displayName',
    'type_id', 'type_name', 'result_id', 'result_name'
]
mov_actions = df_xT[columnas_deseadas].copy()
mov_actions
```

```python
#Se aplica el modelo Expected Threat con un nuevo dataframe
mov_actions = xthreat.get_successful_move_actions(mov_actions)
mov_actions["xT_value"] = xT_model.rate(mov_actions)
mov_actions[["start_x", "start_y", "end_x", "end_y", "xT_value"]]
```

```python
#Dibujamos el campo para ver los xT
# Crear los bins del campo
x_bins = np.linspace(0, 105, 13)  # 12 zonas horizontales
y_bins = np.linspace(0, 68, 9)    # 8 zonas verticales

# Crear matriz vacía para almacenar valores xT promedio por zona
heatmap_matrix = np.zeros((len(x_bins)-1, len(y_bins)-1))

# Agrupar por celdas y calcular media de xT
for i in range(len(x_bins)-1):
    for j in range(len(y_bins)-1):
        cell = mov_actions[
            (mov_actions["start_x"] >= x_bins[i]) & (mov_actions["start_x"] < x_bins[i+1]) &
            (mov_actions["start_y"] >= y_bins[j]) & (mov_actions["start_y"] < y_bins[j+1])
        ]
        if not cell.empty:
            avg = cell["xT_value"].mean()
            # Clipping for seguridad: asegurar que esté entre 0 y 1
            heatmap_matrix[i, j] = np.clip(avg, 0, 1)

# Transponer para ajustar a matplotsoccer
heatmap_matrix = heatmap_matrix.T

# Mostrar el mapa de calor
mps.heatmap(
    heatmap_matrix,
    cmap="hot",
    linecolor="white",
    cbar=True
)
```

```python
#Por ultimo en esta medida agregamos Xt value al dataframe original
df['Xt_value'] = mov_actions['xT_value']
```

## Eficiencia ofensiva tras recuperación en campo alto

```python
df['time_seconds'] = pd.to_numeric(df['time_seconds'], errors='coerce')
df['HighTurnover'] = 0
df['HighTurnoverShot'] = 0

acciones_defensivas = ['Interception', 'BallRecovery']

for i in range(len(df)):
        fila = df.iloc[i]

        if (
            fila['type_displayName'] in acciones_defensivas and
            fila['outcomeType_displayName'] == 'Successful' and
            fila['x'] > 50  # Campo rival
        ):
            equipo = fila['teamId']
            tiempo_inicio = fila['time_seconds']
            df.at[i, 'HighTurnover'] = 1

            # Buscar si en los próximos eventos hay un tiro del mismo equipo en los siguientes 20 segundos
            for j in range(i + 1, min(i + 20, len(df))):  # Limita la búsqueda a los 20 siguientes eventos (ajustable)
                siguiente = df.iloc[j]
                if (
                    siguiente['teamId'] == equipo and
                    siguiente['isShot'] == 1 and
                    siguiente['time_seconds'] - tiempo_inicio <= 20
                ):
                    df.at[i, 'HighTurnoverShot'] = 1

                    
```

## Expected goal

```python
# =======================
# 1. Carga EL MODELO
# =======================

with open('Modelos/xg_model.pkl', 'rb') as f:
    model = pickle.load(f)
    
b = model.params.values  # <- ESTO DEFINE b
model_variables = model.model.exog_names[1:]  # omitir el intercepto
# =======================
# 2. PREPARAR TUS DATOS
# =======================

# Renombrar columnas según tu estructura de eventos
rename_dict = {
    'type_value': 'eventId',
    'playerId': 'playerId',
    'matchId': 'matchId',
    'Shot': 'eventName',
    'teamId': 'teamId',
    'period_value': 'matchPeriod',
    'Segundos': 'eventSec',
    'id': 'id',
    'x': 'X',
    'y': 'Y'
}

df_xG = df.rename(columns=rename_dict).copy()
df_xG = df_xG[df_xG['eventName'] == 1]  # Filtrar solo disparos

# =======================
# 3. CALCULAR xG
# =======================

# Coordenadas ajustadas
df_xG["X"] = (100 - df_xG['X']) * 105 / 100
df_xG["Y"] = df_xG['Y'] * 68 / 100
df_xG["C"] = abs(df_xG['Y'] - 50) * 68 / 100

# Distancia y ángulo
df_xG["Distance"] = np.sqrt(df_xG["X"]**2 + df_xG["C"]**2)
df_xG["Angle"] = np.where(
    np.arctan(7.32 * df_xG["X"] / (df_xG["X"]**2 + df_xG["C"]**2 - (7.32/2)**2)) > 0,
    np.arctan(7.32 * df_xG["X"] / (df_xG["X"]**2 + df_xG["C"]**2 - (7.32/2)**2)),
    np.arctan(7.32 * df_xG["X"] / (df_xG["X"]**2 + df_xG["C"]**2 - (7.32/2)**2)) + np.pi
)

# Variables del modelo
df_xG["X2"] = df_xG['X']**2
df_xG["C2"] = df_xG['C']**2
df_xG["AX"] = df_xG['Angle'] * df_xG['X']

# Función para calcular xG
def calculate_xG(row):    
    bsum = b[0]  # Intercepto
    for i, v in enumerate(model_variables):
        bsum += b[i+1] * row[v]
    xG = 1 / (1 + np.exp(-bsum))  # Regresión logística
    return xG

# Aplicar al DataFrame
df["xG"] = df_xG.apply(calculate_xG, axis=1)# Regresión logística
df
```
