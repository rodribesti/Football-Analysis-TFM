```python
# train_xg_model.py
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pickle
import os

# Cargar datos de entrenamiento
shots = pd.read_excel(r'G:\Mi unidad\TFM\OPTA\Modelos\Eventes_Premier_train.xlsx')

# Crear variables
shots["X2"] = shots['X']**2
shots["C2"] = shots['C']**2
shots["AX"] = shots['Angle'] * shots['X']

model_variables = ['Angle', 'Distance', 'X', 'C', 'X2', 'C2', 'AX']
model_formula = "Goal ~ " + ' + '.join(model_variables)

# Entrenar modelo
model = smf.glm(formula=model_formula, data=shots, family=sm.families.Binomial()).fit()

# Guardar modelo
os.makedirs("modelos", exist_ok=True)
with open("Modelos/xg_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("✅ Modelo entrenado y guardado.")
```
