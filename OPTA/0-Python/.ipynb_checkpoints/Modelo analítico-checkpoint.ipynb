## Introducción

#### En el presente notebook se establece la conexión con la base de datos alojada en Azure, la cual alberga tanto los datos en crudo como los datos ya procesados correspondientes a los jugadores de la temporada 2024 de la Major League Soccer (MLS). A partir de esta información, se procederá a la elaboración y cálculo de métricas avanzadas de rendimiento (calculo de percentiles), que servirán como base para su posterior visualización y análisis en Power BI.

```python
import json
import pandas as pd
import pyodbc
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Ruta al archivo JSON
config_path = r'G:\Mi unidad\TFM\OPTA\0-Python\config\config_bbdd.json'

# Leer archivo de configuración
with open(config_path, 'r') as f:
    config = json.load(f)

# Acceder a los datos dentro de la clave 'db_TFM'
db_config = config['db_TFM']

# Establecer los parámetros
server = db_config['host']
database = db_config['database']
username = db_config['user']
password = db_config['password']
driver = db_config['driver']
# puerto no se usa explícitamente en pyodbc para Azure SQL

# Cadena de conexión ODBC
conn_str = f"""
    DRIVER={{{driver}}};
    SERVER={server},1433;
    DATABASE={database};
    UID={username};
    PWD={password};
    Encrypt=yes;
    TrustServerCertificate=no;
    Connection Timeout=30;
"""

# Ejecutar la consulta
try:
    conn = pyodbc.connect(conn_str)
    query = "SELECT * FROM dbo.FAC_JUGADOREVENTO"
    df = pd.read_sql(query, conn)
    print("✅ Conexión exitosa. Datos leídos correctamente.")
    display(df.head())  # si usas Jupyter
except Exception as e:
    print(f"❌ Error al conectar o ejecutar la consulta: {e}")
finally:
    conn.close()
```

```python
df.columns
```

```python
columnas_percentiles = [
    'ShortPass_per90', 'ShortPass_per_possession',
    'KeyPasses_per90', 'KeyPasses_per_possession',
    'ProgressPasses_per90', 'ProgressPasses_per_possession',
    'TouchesRivalArea_per90', 'TouchesRivalArea_per_possession',
    'TouchesOwnArea_per90', 'TouchesOwnArea_per_possession',
    'DribbleSuccess_per90', 'DribbleSuccess_per_possession',
    'DribbleFail_per90', 'DribbleFail_per_possession',
    'DefActionSuccess_per90', 'DefActionSuccess_per_possession',
    'DefActionFail_per90', 'DefActionFail_per_possession',
    'EntriesSuccess_per90', 'EntriesSuccess_per_possession',
    'AerialSuccess_per90', 'AerialSuccess_per_possession',
    'AerialFail_per90', 'AerialFail_per_possession',
    'DuelsSuccess_per90', 'DuelsSuccess_per_possession',
    'DuelsFail_per90', 'DuelsFail_per_possession',
    'Clearance_per90', 'Clearance_per_possession',
    'DefensiveToTransition_per90', 'DefensiveToTransition_per_possession',
    'HighTurnover_per90', 'HighTurnover_per_possession',
    'HighTurnoverShot_per90', 'HighTurnoverShot_per_possession',
    'avg_xG', 'avg_xT'
]
```

```python
for col in columnas_percentiles:
    if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):
        df[f'{col}_percentil'] = df[col].rank(pct=True) * 100
        df[f'{col}_percentil'] = df[f'{col}_percentil'].round(2)
    else:
        print(f"⚠️  La columna '{col}' no está en el DataFrame o no es numérica.")

# 4. Guardar para Power BI (opcional)
print("✅ Percentiles calculados y exportados correctamente.")
df.shape
```

```python
columna_metrica = 'ShortPass_per90_percentil'
columna_nombre = 'playerName'
jugador_objetivo = 'N. Palacio'

# Obtener valor del jugador
valores_jugador = df[df[columna_nombre] == jugador_objetivo][columna_metrica].values
if len(valores_jugador) == 0:
    raise ValueError(f"No se encontró el jugador {jugador_objetivo} en la columna {columna_metrica}")
valor_jugador = valores_jugador[0]

# Estadísticas
media = df[columna_metrica].mean()
std = df[columna_metrica].std()

# Rango fijo para eje X
x_min, x_max = 0, 200
x_vals = np.linspace(x_min, x_max, 100)
pdf_vals = norm.pdf(x_vals, media, std)

plt.figure(figsize=(10, 6))
plt.fill_between(x_vals, pdf_vals, color='skyblue', alpha=0.7, label='Distribución normal teórica')
plt.plot(x_vals, pdf_vals, color='black', linestyle='--', linewidth=2)
plt.axvline(valor_jugador, color='red', linestyle='--', linewidth=2, label=f'{jugador_objetivo}')
plt.xlim(x_min, x_max)
plt.title(f'Distribución normal teórica de {columna_metrica}')
plt.xlabel(columna_metrica)
plt.ylabel('Densidad')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Lista de columnas donde quieres cambiar '.' por ','
cols = [
    'total_minutos_jugados', 'porcentaje_medio_posesion',
    'total_short_pass', 'ShortPass_per90', 'ShortPass_per_possession',
    'total_key_passes', 'KeyPasses_per90', 'KeyPasses_per_possession',
    'total_progress_passes', 'ProgressPasses_per90',
    'ProgressPasses_per_possession', 'total_touches_rival',
    'TouchesRivalArea_per90', 'TouchesRivalArea_per_possession', 'avg_xG',
    'total_touches_own', 'TouchesOwnArea_per90',
    'TouchesOwnArea_per_possession', 'avg_xT', 'total_dribble_success',
    'DribbleSuccess_per90', 'DribbleSuccess_per_possession',
    'total_dribble_fail', 'DribbleFail_per90', 'DribbleFail_per_possession',
    'total_def_action_success', 'DefActionSuccess_per90',
    'DefActionSuccess_per_possession', 'total_def_action_fail',
    'DefActionFail_per90', 'DefActionFail_per_possession',
    'total_entries_success', 'EntriesSuccess_per90',
    'EntriesSuccess_per_possession', 'total_aerial_success',
    'AerialSuccess_per90', 'AerialSuccess_per_possession',
    'total_aerial_fail', 'AerialFail_per90', 'AerialFail_per_possession',
    'total_duels_success', 'DuelsSuccess_per90',
    'DuelsSuccess_per_possession', 'total_duels_fail', 'DuelsFail_per90',
    'DuelsFail_per_possession', 'total_clearance', 'Clearance_per90',
    'Clearance_per_possession', 'total_def_to_trans',
    'DefensiveToTransition_per90', 'DefensiveToTransition_per_possession',
    'total_high_turnover', 'HighTurnover_per90',
    'HighTurnover_per_possession', 'total_high_turnover_shot',
    'HighTurnoverShot_per90', 'HighTurnoverShot_per_possession'
]

# Convertir a string y reemplazar '.' por ',' para las columnas indicadas
for col in cols:
    if col in df.columns:
        # Aseguramos que sea string para el replace
        df[col] = df[col].astype(str).str.replace('.', ',', regex=False)
    else:
        print(f"Columna {col} no encontrada en el DataFrame.")

# Guardar como CSV (el parámetro decimal=',' es para que pandas guarde coma como decimal en números,
# pero aquí ya cambiamos manualmente el punto por coma en texto, así que el csv se guarda tal cual)
df.to_csv('FAC_MODELO_ANALÍTICO.csv', index=False, encoding='utf-8-sig')

print("Archivo CSV guardado como 'FAC_MODELO_ANALÍTICO.csv'")
```
