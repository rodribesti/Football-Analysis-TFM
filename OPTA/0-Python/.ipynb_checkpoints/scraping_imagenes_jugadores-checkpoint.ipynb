```python
def extraer_jugadores_besoccer(url):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    players_data = []

    for td in soup.find_all("td", class_="player-img"):
        a_tag = td.find("a", href=True)
        img_tag = td.find("img")

        if a_tag and img_tag:
            perfil_url = a_tag["href"]
            nombre = perfil_url.strip("/").split("/")[-1].replace("-", " ").title()
            imagen_url = img_tag.get("src")

            players_data.append({
                "nombre": nombre,
                "imagen_url": imagen_url
            })

    return pd.DataFrame(players_data)
```

```python
df_imagenes = pd.read_excel(r'G:\Mi unidad\TFM\OPTA\0-Python\jugadores_con_imagenes_asignadas.xlsx')
df_imagenes
```

```python
# Asegúrate de que imagen_url es tipo string (incluso si hay NaN)
df_imagenes['imagen_url'] = df_imagenes['imagen_url'].astype(str)
df_unicos = df_imagenes.drop_duplicates(subset='playerId')
```

```python
# Leer configuración
bdd = "db_TFM"
with open('config/config_bbdd.json', 'r') as file:
    config_data = json.load(file)

server = config_data[bdd]["host"]
db = config_data[bdd]["database"]
u = config_data[bdd]["user"]
p = config_data[bdd]["password"]

# Cadena de conexión
conn_str = (
    "DRIVER=ODBC Driver 18 for SQL Server;"
    f"SERVER={server};"
    f"DATABASE={db};"
    f"UID={u};"
    f"PWD={p};"
    "TrustServerCertificate=yes;"
)

# Conectar y crear cursor
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()


# Insertar datos en la tabla
for index, row in df_unicos.iterrows():
    cursor.execute("""
        INSERT INTO jugadores_imagenes (playerId, imagen_url)
        VALUES (?, ?)
    """, row['playerId'], row['imagen_url'])

# Confirmar cambios y cerrar conexión
conn.commit()
cursor.close()
conn.close()

print("✅ Datos añadidos correctamente a la tabla 'jugadores_imagenes'.")
```
