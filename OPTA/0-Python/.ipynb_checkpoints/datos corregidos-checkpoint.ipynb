```python
import pandas as pd
# Ruta del archivo CSV
ruta = r"G:\Mi unidad\TFM\OPTA\0-Python\FAC_EVENT.csv"

# Leer el CSV
df = pd.read_csv(ruta, sep=';')

# Mostrar las primeras filas para comprobar
df.columns
```

```python
columnas_deseadas = ['id', 'matchId', 'playerId', 'xG', 'xT_value']
df_filtrado = df[columnas_deseadas]
```

```python
# Cambiar el tipo de las columnas matchId y playerId a string (object en pandas)
df_filtrado['matchId'] = df_filtrado['matchId'].astype(str)
df_filtrado['playerId'] = df_filtrado['playerId'].astype(str)
```

```python
print(df_filtrado.dtypes)
```

```python
# Guardar df_filtrado en CSV
ruta_csv = r'G:\Mi unidad\TFM\OPTA\0-Python\data_filtrado.csv'
df_filtrado.to_csv(ruta_csv, index=False, encoding='utf-8-sig')

print(f'DataFrame guardado en {ruta_csv}')
```
