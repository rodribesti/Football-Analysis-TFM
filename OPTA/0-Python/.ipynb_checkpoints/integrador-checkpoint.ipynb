# CSV ---> BBDD

```python
ruta_dest = r"G:\Mi unidad\TFM\OPTA\0-Python\scoresway"
# Ruta de la carpeta que contiene los archivos
folder_path =  os.path.join(os.getcwd(),"{}/output".format(ruta_dest))  # Esto devuelve la ruta en la que están los ficheros planos
```

```python
# Define aquí manualmente los valores que estás usando
season = "2023-2024"
comp = "MLS Next Pro"

ficheros = [f for f in os.listdir(os.path.join(os.getcwd(),"fixtures")) if f.endswith(('.xlsx', '.xls')) and "$" not in f]
fixtures = pd.DataFrame()
for i in ficheros:
    fi = pd.read_excel(os.path.join(os.getcwd(),"fixtures",i))
    fixtures=pd.concat([fixtures,fi])
fixtures = fixtures[(fixtures.season==season) & (fixtures.competition==comp)]
```

```python
fixtures['matchData'] = fixtures.match_id.apply(lambda x: x+"_matchData.csv")
fixtures['teamStats'] = fixtures.match_id.apply(lambda x: x+"_teamStats.csv")
fixtures['teamData'] = fixtures.match_id.apply(lambda x: x+"_teamData.csv")
fixtures['playerData'] = fixtures.match_id.apply(lambda x: x+"_playerData.csv")
fixtures['eventData'] = fixtures.match_id.apply(lambda x: x+"_eventData.csv")
```

# 1. MATCH_DATA

```python
pattern = os.path.join(folder_path, '*matchData*.csv')
file_list = glob.glob(pattern)
file_list = [i for i in file_list if i.split("\\")[-1] in list(fixtures.matchData.unique())]
print("Archivos encontrados:", file_list)
```

```python
existing_games = pd.read_sql("SELECT matchId FROM match_data", con=engine)
existing_games_set = list(existing_games['matchId'].unique())
```

```python
data=pd.DataFrame()
if not file_list:
    print("No se encontraron archivos que cumplan con el patrón en la carpeta actual.")
else:
    for file in file_list:
        df = pd.read_csv(file, sep=';', engine='python')
        
        data=pd.concat([data,df])
        print(f"Archivo {os.path.basename(file)} cargado.")

if data.shape[0]>0:
    data = data[~data.matchId.isin(existing_games_set)]
    data['origen']=ruta_dest
    data.to_sql(name='match_data', con=engine, if_exists='append', index=False)
    print("¡Todos los archivos han sido añadidos a la tabla match_data!")
```

# 6. EVENT_DATA

```python
pattern = os.path.join(folder_path, '*eventData*.csv')
file_list = glob.glob(pattern)
file_list = [i for i in file_list if i.split("\\")[-1] in list(fixtures.eventData.unique())]
print("Archivos encontrados:", file_list)
```

```python
folder_path = 'G:/Mi unidad/TFM/OPTA/0-Python/scoresway/output'
file_list = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('eventData.csv')]

if not file_list:
    print("No se encontraron archivos que cumplan con el patrón en la carpeta actual.")
else:
    for file in file_list:
        try:
            df = pd.read_csv(file, decimal=",", sep=';', engine='python')

            for col in df.columns:
                if df[col].dtype == 'object':
                    df[col] = df[col].str.replace(',', '.', regex=False)
                    df[col] = pd.to_numeric(df[col], errors='ignore')

            df = df.drop_duplicates(subset=['id', 'eventId', 'matchId'], keep='first')

            with engine.begin() as conn:
                conn.execute(
                    text("DELETE FROM event_data WHERE matchId = :match_id"),
                    {"match_id": df.matchId.values[0]}
                )

            df.to_sql(name='event_data', con=engine, if_exists='append', index=False, method='multi', chunksize=5000)
            print(f"Archivo {os.path.basename(file)} cargado.")
        
        except IntegrityError as e:
            print(f"Error de integridad al cargar {file}: {e}")
        except Exception as e:
            print(f"Otro error al cargar {file}: {e}")

print("¡Todos los archivos han sido añadidos a la tabla event_data!")
```
